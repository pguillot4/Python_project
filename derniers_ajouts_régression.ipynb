On s'intéresse ensuite à l'effet différencié de la variable taux_sup, dans le cas des lycées privés et des lycées publics.

```
df_final_pub = df_final[df_final["Secteur_public"] == 1]
df_final_priv = df_final[df_final["Secteur_public"] == 0]

Y_pub = df_final_pub["Taux_reussite_S"]
X_pub = df_final_pub[["Taux_sup"]]
X_pub = sm.add_constant(X_pub)
model = sm.OLS(Y_pub, X_pub)
ols_model = model.fit(robust = 'hc3')
print(ols_model.summary())


Y_priv = df_final_priv["Taux_reussite_S"]
X_priv = df_final_priv[["Taux_sup"]]
X_priv = sm.add_constant(X_priv)
model = sm.OLS(Y_priv, X_priv)
ols_model = model.fit(robust = 'hc3')
print(ols_model.summary())
```

On constate que dans les deux modèles, le R2 est très faible (1.7 % dans le cas des lycées publics, et 4% dans le cas des lycées privés). La variable **taux_sup** ne permet donc pas à elle seule d'expliquer la variabilité des taux de réussite au bac S par lycée. On remarque cependant que le coefficient associé à la variable **taux_sup** est plus important dans le cas des lycées privés (0.0831), que dans la base des lycées publics (0.0710) . On observe graphiquement ce résultat en comparant l'aspect des deux nuages de points ci-dessus : la droite semble plus pentue sur le graphique de droite que sur celui de gauche.
